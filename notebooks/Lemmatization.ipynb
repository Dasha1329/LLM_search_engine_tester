```python
%%time
import swifter
from pymorphy3 import MorphAnalyzer
from nltk.tokenize import word_tokenize

pymorphy3_analyzer = MorphAnalyzer()

def lemmatize_text(text):
    if pd.isnull(text):
        return ''
    tokens = word_tokenize(text.lower())  # Разбиваем текст на токены
    lemmas = [pymorphy3_analyzer.parse(token)[0].normal_form for token in tokens]  # Лемматизация
    return ' '.join(lemmas)  # Объединяем токены обратно в строку

# df['lemma_keyword'][:1000] = df['keyword'][:1000].apply(lemmatize_text)

# df['lemma_keyword'][:1000] = df['keyword'][:1000].swifter.apply(lemmatize_text)


from tqdm.auto import tqdm
tqdm.pandas()
```

```python
keywords_list = df1['keyword'].tolist()
# map()
# lemma_keyword_list = list(tqdm(map(lemmatize_text, keywords_list), desc='Processing quiries', total=len(keywords_list)))

# list comprehension [] 
lemma_keyword_list = [lemmatize_text(query) for query in tqdm(keywords_list, desc="Processing queries")]

# df1['lemma_keyword'] = df1['keyword'].progress_apply(lemmatize_text) 
```

```python
lemma_keyword_list
```

```python
df1['lemma_keywords'] = lemma_keyword_list
```

# Стемминг (окончания) не делала!

```python
import nltk
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
```

```python
stemmer = SnowballStemmer("russian")
def stem_text(text):
    tokens = word_tokenize(text.lower())  # Токенизация и перевод в нижний регистр
    stems = [stemmer.stem(token) for token in tokens]  # Стемминг каждого токена
    return ' '.join(stems)  # Объединение токенов обратно в строку
```

```python
from tqdm.auto import tqdm
tqdm.pandas()

keywords_list = df1['keyword'].tolist()
# map()
# lemma_keyword_list = list(tqdm(map(lemmatize_text, keywords_list), desc='Processing quiries', total=len(keywords_list)))

# list comprehension [] 
stem_keyword_list = [stem_text(query) for query in tqdm(keywords_list, desc="Processing queries")]
```

```python
stem_keyword_list
```

```python
df1['stem_keyword'] = stem_keyword_list
df1
```

```python
df1.to_parquet('stem_keyword_df3', index=False, engine='pyarrow')
```

# Выделение низкочастотных запросов

```python
keyword_counts = df.groupby(['keyword', 'user_id']).size()
keyword_counts
```

```python
keyword_counts = keyword_counts.to_frame().reset_index()
```

```python
keyword_counts.columns = ['keyword', 'user_id', 'frequency']
keyword_counts
```

```python
# нижние 2%
low_freq_threshold = keyword_counts['frequency'].quantile(0.02)
print(low_freq_threshold)
low_freq_keywords = keyword_counts[keyword_counts['frequency'] <= low_freq_threshold]

print(f"Количество низкочастотных запросов: {len(low_freq_keywords)}")
```
